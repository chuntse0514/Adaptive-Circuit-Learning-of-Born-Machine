# Discretization and truncation of continuous distribution

This notebook closely follows the discretized distribution presented in the reference paper: DOI 10.1186/s40488-015-0028-6

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Discrete normal distribution  </br>

$$f(x)=\frac{1}{\sqrt{2\pi\sigma^2}}\exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$$

```python
x = np.linspace(-32, 31, 1024)

sigma = 4
mu = 5

k = np.linspace(-32, 31, 64)
lambda_ = np.exp((2*mu-1)/(2*sigma**2))
q = np.exp(-1/sigma**2)
p = (lambda_ ** k) * q ** (k * (k-1) / 2)
p = p / np.sum(p)

f = 1 / np.sqrt(2 * np.pi * sigma ** 2) * np.exp(-(x-mu)**2 / (2 * sigma**2) )
plt.plot(x, f, color='black', label='PDF', ls='--', alpha=0.3)
plt.bar(k, p, color='orange', label='PMF (64)')
plt.legend()
plt.ylim(bottom=0)
plt.show()
```

## Discrete exponential

```python
x = np.linspace(0, 63, 1024)
k = np.arange(64)

lambda_ = 0.1
f = lambda_ * np.exp(-lambda_ * x)

p = (1 - np.exp(-lambda_)) * np.exp(-lambda_ * k)
p = p / np.sum(p)

plt.plot(x, f, color='black', label='PDF', ls='--', alpha=0.3)
plt.bar(k, p, color='yellowgreen', label='PMF (64)')
plt.legend()
plt.show()
```

## Discrete log normal 
$$
f(x)=\frac{1}{x\sigma\sqrt{2\pi}}\exp\left(-\frac{(\ln x-\mu)^2}{2\sigma^2}\right) 
\\[1cm]
p(k)=e^{-\frac{(\ln k-\mu}{2\sigma)^2}} / k\sum_{j=1}^{\infty}e^{-\frac{(\ln j-\mu)^2}{2\sigma^2}}, \quad\quad k=1, 2, ...
$$

```python
x = np.linspace(1, 64, 1024)
k = np.arange(64) + 1

mu = 3
sigma = 0.3

f = 1 / (x * sigma * np.sqrt(2 * np.pi)) * np.exp(-(np.log(x)-mu)**2 / (2 * sigma**2))
p = np.exp(-(np.log(k)-mu) ** 2 / (2 * sigma**2)) / k 
p = p / np.sum(p)

plt.plot(x, f, label='PDF', color='black', ls='--', alpha=0.3)
plt.bar(k, p, label='PMF (64)', color='turquoise')
plt.legend()
plt.show()
```
