```python
import json
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import matplotlib.gridspec as gridspec
from data import *

def plot_reduce_pool(file_0, file_1):
    with open(file_0, 'r') as loss_0:
        kl_div_0 = json.load(loss_0)['kl div']['epoch']
    with open(file_1, 'r') as loss_1:
        kl_div_1 = json.load(loss_1)['kl div']['epoch']

    plt.semilogy(np.arange(len(kl_div_0))+1, kl_div_0, ls='-', color='blue', label='normal $(N_o=400)$', alpha=0.5)
    plt.semilogy(np.arange(len(kl_div_1))+1, kl_div_1, ls='-', color='orange', label='r=0.5 $(N_o=175)$', alpha=0.5)
    
    plt.semilogy([len(kl_div_0)], kl_div_0[-1], marker='X', markersize=10)
    plt.semilogy([len(kl_div_1)], kl_div_1[-1], marker='X', markersize=10)

    plt.legend()
    plt.grid(alpha=0.75, ls=":")
    plt.xlabel('epoch')
    plt.ylabel('KL divergence')
    plt.title('Reduced pool for bas 4x4')

    plt.savefig('./images/reduced_pool_bas_4x4.png')
```

```python
plot_reduce_pool(
    file_0='./results/ACLBM/ACLBM(data=bas 4x4, No=400, t1=0.0005, t2=0.0005).json',
    file_1='./results/ACLBM/ACLBM(data=bas 4x4, No=175, t1=0.001, t2=0.005, r=0.5).json',
)
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt

def plot_different_No(file_1, file_2, file_3, data):
    with open(file_1, 'r') as loss_1, \
         open(file_2, 'r') as loss_2, \
         open(file_3, 'r') as loss_3:

        kl_div_1 = json.load(loss_1)['kl div']['epoch']
        kl_div_2 = json.load(loss_2)['kl div']['epoch']
        kl_div_3 = json.load(loss_3)['kl div']['epoch']

        No_list = [1, 3, 5] if data == 'real image 1' else [4, 40, 400]
        
        plt.semilogy(np.arange(len(kl_div_1))+1, kl_div_1, ls=':', color='lightcoral', label=f'$N_o={No_list[0]}$')
        plt.semilogy(np.arange(len(kl_div_2))+1, kl_div_2, ls=':', color='red', label=f'$N_o={No_list[1]}$')
        plt.semilogy(np.arange(len(kl_div_3))+1, kl_div_3, ls=':', color='brown', label=f'$N_o={No_list[2]}$')

        plt.legend()
        plt.xlabel('iteration')
        plt.ylabel('KL divergence')
        if data == 'real image 1':
            plt.title('Different $N_o$ for real image 1')
        else:
            plt.title('Different $N_o$ for bas 4x4')
        plt.grid()

        plt.show()
```

```python
plot_different_No(
        file_1='./results/ACLBM/ACLBM-Fisher-Rao(data=real image 1, No=1, t1=0.001, t2=0.005).json',
        file_2='./results/ACLBM/ACLBM-Fisher-Rao(data=real image 1, No=3, t1=0.001, t2=0.005).json',
        file_3='./results/ACLBM/ACLBM-Fisher-Rao(data=real image 1, No=5, t1=0.001, t2=0.005).json',
        data='real image 1'
)
```
