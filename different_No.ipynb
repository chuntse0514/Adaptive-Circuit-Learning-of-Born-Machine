```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

def find_indices(A, B):
    return np.where(np.in1d(A, B))[0]

def plot_different_No(file_1, file_2, file_3):
    with open(file_1 + '.pkl', 'rb') as loss_1, \
        open(file_2 + '.pkl', 'rb') as loss_2, \
        open(file_3 + '.pkl', 'rb') as loss_3:

        loss_history_1, js_div_1, kl_div_1 = pickle.load(loss_1)
        loss_history_2, js_div_2, kl_div_2 = pickle.load(loss_2)
        loss_history_3, js_div_3, kl_div_3 = pickle.load(loss_3)

        kl_indices_1 = find_indices(kl_div_1['iteration'], kl_div_1['epoch'])
        kl_indices_2 = find_indices(kl_div_2['iteration'], kl_div_2['epoch'])
        kl_indices_3 = find_indices(kl_div_3['iteration'], kl_div_3['epoch'])

        plt.plot(np.arange(len(kl_div_1['iteration']))+1, kl_div_1['iteration'], ls=':', color='lightcoral', label='$N_o=3$')
        plt.plot(kl_indices_1+1, kl_div_1['epoch'], ls='', color='lightcoral', marker='x', markersize=5)
        plt.plot(np.arange(len(kl_div_2['iteration']))+1, kl_div_2['iteration'], ls=':', color='red', label='$N_o=30$')
        plt.plot(kl_indices_2+1, kl_div_2['epoch'], ls='', color='red', marker='x', markersize=5)
        plt.plot(np.arange(len(kl_div_3['iteration']))+1, kl_div_3['iteration'], ls=':', color='brown', label='$N_o=300$')
        plt.plot(kl_indices_3+1, kl_div_3['epoch'], ls='', color='brown', marker='x', markersize=5)

        plt.legend()
        plt.xlabel('iteration')
        plt.ylabel('KL divergence')
        plt.title('Different $N_o$ for bas 4x4')
        # plt.yticks(np.linspace(0, 0.1, 11))
        plt.grid()
        plt.ylim(bottom=0)

        plt.show()
```

```python
plot_different_No(
        file_1='./results/ACLBM(data=real image 1, loss=KL divergence, Ng=1, t1=0.001, t2=0.005)',
        file_2='./results/ACLBM(data=real image 1, loss=KL divergence, Ng=3, t1=0.001, t2=0.005)',
        file_3='./results/ACLBM(data=real image 1, loss=KL divergence, Ng=5, t1=0.001, t2=0.005)'
)
```

```python
plot_different_No(
        file_1='./results/ACLBM(data=bas 4x4, loss=KL divergence, Ng=3, t1=0.001, t2=0.005)',
        file_2='./results/ACLBM(data=bas 4x4, loss=KL divergence, Ng=30, t1=0.001, t2=0.005)',
        file_3='./results/ACLBM(data=bas 4x4, loss=KL divergence, Ng=300, t1=0.001, t2=0.005)'
)
```
